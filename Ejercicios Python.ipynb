```python
# Solicitar el valor del radio de un circulo para calcular su área:

from math import pi

radio = float(input('Ingrese el radio: '))

area = pi*(radio**2)

print("El área del circulo es: ", area)
```

```python
# Obtener la representación inversa de una cadena de caracteres

cadena = 'python'

# Primera Forma
r = cadena[::-1]
print(r)
print("\n")

# Segunda Forma
for i in range(len(cadena)-1, -1, -1):
  print (cadena[i], end='')
```

```python
# Crear una función para evaluar un numero y realizar una operación
#si n<=15, entonces 15 - n; (15 - n)*2

n = float(input('Ingrese un numero: '))

def evaluar (n):
  if n<=15:
    return 15-n
  else:
    return (15-n)*2

print ('la suma es: ', evaluar(n))
```

```python
# Genere los "n" primeros numeros pares positivos

n = int (input("Ingrese un número: "))
p = int (input("Ingrese el número de pares: "))

for i in range(1,n):
  if i%2==0 & i<p:
    print(p)
  else:
    continue
```
