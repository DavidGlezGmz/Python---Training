**4. Operadores** 

Calcularemos el área de un triangulo para aprender los operadores básicos.

También el promedio de las distintas notas.

```python
# Hallar el área de un triangulo
''' 
DATOS: 
B = Base
H = Altura
Area = (Base * Altura) / 2
'''
B = float (input('Ingrese la base: '))
H = float (input('Ingrese la altura: '))
A = (B*H)/2
print("El resultado del area del triangulo es: ", A)
```

```python
# Hallar el promedio de tres notas
''' 
DATOS: 
N1 = Nota 1
N2 = Nota 2
N3 = Nota 3
P = Promedio
'''
N1 = float(input('Ingrese la nota 1: '))
N2 = float(input('Ingrese la nota 2: '))
N3 = float(input('Ingrese la nota 3: '))
P = (N1 + N2 + N3)/3
print("El promedio de las notas es: ", P)
```

**8. Estructuras Iterativas FOR / WHILE** 

Muchas veces, es necesario repetir una instrucción ya que así lo requiere el algoritmo. Para no repetir el codigo más de una vez, en estos casos usaremos la instrucción ciclo.

WHILE: El bloque de la secuencia de instrucciones se repite siempre que la condicion dada por el valor de una variable (Si la expresion logica es verdadera)

FOR: Se caracteriza por repetirse un numero determinado de veces.

```python
#Suma de los números del 1 al 6

N = 6
Suma = 0
i = 1

while i <= N:
  Suma = Suma + i
  i = i + 1
  print ("La suma de los numeros hasta ", N, "es: ", Suma)
print ("\nLa suma de los numeros hasta ", N, "es: ", Suma)
```

**11. Break / Continue** 

La instrucción continue se usa dentro de un ciclo FOR asociado a un ciclo SI y no ejecuta la linea de código.

```python
# Identificar pares e impares

N1 = int(input("Escriba un numero: "))
N2 = int(input(f"Escriba un numero mayor o igual que {N1}: "))

if N2<=N1:
  print (f"Ingrese un numero mayor o igual que {N1}")
else:
  for i in range (N1, N2+1):
    if i%2 == 0: 
      print (f"El numero {i} par")
    else: 
      print (f"El numero {i} impar")
```

```python
# Identificar pares

N1 = int(input("Escriba un numero: "))

for i in range (1, N1+1):
  if i%2 == 0: 
    print (f"El numero {i} par")
```

```python
# Identificar impares

N1 = int(input("Escriba un numero: "))

for i in range (1, N1+1):
  if i%2 == 0: 
    continue # SALTA DIRECTO AL ELSE
    print (f"El numero {i} par")
  else: 
    print (f"El numero {i} impar")
```

```python
# Identificar pares

N1 = int(input("Escriba un numero: "))

for i in range (1, N1+1):
  if i%2 == 0: 
    break #Rompe, sale del ciclo
    print (f"El numero {i} par")
  else: 
    print (f"El numero {i} impar")
```

```python
#Seleccionar los impares y romper cuando encuentre un numero divisible por 6

N = int(input("Ingrese un número: "))

for i in range (1, N):
  if (i%2==0):
    continue
    print (i)
  else: 
    print (i)
    if ((i+1)%6 == 0):
      break
```

```python
# Separar una lista en pares e impares

ejemplo = [1,4,8,7,2,9,12,21]

def separar (lista):
  lista.sort()
  par = []
  impar = []

  for i in lista:
    if i%2 == 0:
      par.append(i)
    else:
      impar.append(i)
  return par, impar

par, impar = separar(ejemplo)
print ("Los elementos pares son: ", par)
print ("Los elementos impares son: ", impar)
```

```python
# Calcular 10 numeros aleatorios

import random 

def vector_aleatorio(n):
  vector = [0]*n
  for i in range (n):
    vector [i] = random.randint(0,10)
  return vector

print ("Cuantos numeros aleatorios desea tener: ")
n = int(input())

aleatorio = vector_aleatorio(n)
print (aleatorio)
```
